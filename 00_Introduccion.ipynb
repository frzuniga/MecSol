## Módulo 2 — Tensores de Esfuerzo y Deformación

Los tensores son el lenguaje fundamental de la mecánica del continuo.  
**Prerrequisito:** Álgebra Matricial.

---

### 📓 [Análisis del Tensor de Esfuerzos](Análisis%20del%20Tensor%20de%20Esfuerzos.ipynb)

**Qué aprenderás:** Calcular y visualizar el estado de esfuerzo en un punto.  
**Conceptos clave:**
$$\boldsymbol{\sigma} = \begin{bmatrix} \sigma_{xx} & \tau_{xy} & \tau_{xz} \\ \tau_{xy} & \sigma_{yy} & \tau_{yz} \\ \tau_{xz} & \tau_{yz} & \sigma_{zz} \end{bmatrix}$$
- **Esfuerzos principales** $\sigma_1 \geq \sigma_2 \geq \sigma_3$: eigenvalores de $\boldsymbol{\sigma}$
- **Direcciones principales**: eigenvectores (siempre ortogonales para tensores simétricos)
- **Esfuerzo cortante máximo**: $\tau_{max} = (\sigma_1 - \sigma_3)/2$
- **Invariantes**: $I_1 = \text{tr}(\boldsymbol{\sigma})$, $I_3 = \det(\boldsymbol{\sigma})$

**Conexión:** Los esfuerzos principales son la entrada del Círculo de Mohr y los criterios de fluencia.

---

### 📓 [Matriz de Def. (Gradiente de Deformación)](Matriz%20de%20Def.ipynb)

**Qué aprenderás:** Entender cómo se describe geométricamente la deformación de un sólido continuo.  
**Conceptos clave:**
- **Tensor gradiente de deformación** $\mathbf{F}$: $F_{ij} = \partial x_i / \partial X_j$
- **Matriz de gradiente de desplazamiento**: $\mathbf{J} = \mathbf{F} - \mathbf{I}$
- **Tensor de deformaciones infinitesimales** (pequeñas deformaciones):
$$\boldsymbol{\varepsilon} = \tfrac{1}{2}(\mathbf{F} + \mathbf{F}^T) - \mathbf{I}$$
- **Tensor de Green-Lagrange** (grandes deformaciones):
$$\mathbf{E} = \tfrac{1}{2}(\mathbf{F}^T\mathbf{F} - \mathbf{I})$$

**Conexión:** El tensor $\boldsymbol{\varepsilon}$ es la base del Análisis del Tensor de Deformación y la Ley de Hooke.

---

### 📓 [Análisis del Tensor de Deformación](Análisis%20del%20Tensor%20de%20Deformación.ipynb)

**Qué aprenderás:** Caracterizar el estado de deformación en un punto.  
**Conceptos clave:**
$$\boldsymbol{\varepsilon} = \begin{bmatrix} \varepsilon_{xx} & \varepsilon_{xy} & \varepsilon_{xz} \\ \varepsilon_{xy} & \varepsilon_{yy} & \varepsilon_{yz} \\ \varepsilon_{xz} & \varepsilon_{yz} & \varepsilon_{zz} \end{bmatrix}$$
- **Deformaciones principales**: eigenvalores de $\boldsymbol{\varepsilon}$
- **Deformación volumétrica**: $\varepsilon_v = \text{tr}(\boldsymbol{\varepsilon}) = \varepsilon_1 + \varepsilon_2 + \varepsilon_3$
- Nota: deformaciones ingenieriles $\gamma_{ij} = 2\varepsilon_{ij}$

**Conexión:** El tensor de deformación es la variable de entrada de la Ley de Hooke.

---

### 📓 [Cálculo de la Deformación Volumétrica](cálculo%20de%20la%20Deformación%20Volumétrica.ipynb)  
*(Cuaderno complementario)*

**Qué aprenderás:** Determinar si un material se expande o contrae.  
**Fórmula central:** $\varepsilon_v = \varepsilon_{xx} + \varepsilon_{yy} + \varepsilon_{zz}$

---

### 📓 [Ley de Hooke Generalizada en 3D](Ley%20de%20Hooke%20Generalizada%20en%203D.ipynb)

**Qué aprenderás:** Relacionar esfuerzos y deformaciones para un material elástico isotrópico.  
**Ecuación central:**
$$\boldsymbol{\sigma} = \mathbf{C}\,\boldsymbol{\varepsilon} \qquad \text{con} \qquad \sigma_{ij} = \lambda\delta_{ij}\varepsilon_{kk} + 2\mu\varepsilon_{ij}$$

**Parámetros del material:**

| Símbolo | Nombre | Relación |
|---|---|---|
| $E$ | Módulo de Young | — |
| $\nu$ | Coeficiente de Poisson | $0 \leq \nu \leq 0.5$ |
| $\lambda$ | 1ª constante de Lamé | $\frac{E\nu}{(1+\nu)(1-2\nu)}$ |
| $\mu = G$ | Módulo de corte | $\frac{E}{2(1+\nu)}$ |
| $K$ | Módulo de compresibilidad | $\frac{E}{3(1-2\nu)}$ |

**Rango de aplicabilidad en rocas:** $10^{-6} \leq \varepsilon \leq 10^{-3}$

**Prerrequisito:** Módulo 3.

### 📓 [Tracción Normal y Tangencial en un Plano](Tracción_normal_y_tangencial_a_un%20plano.ipynb)

**Qué aprenderás:** Calcular las fuerzas que actúan sobre planos con orientación arbitraria.  
**Conceptos clave:**
- Vector normal al plano dado rumbo ($\phi$) y echado ($\delta$):
$$\hat{n} = \begin{bmatrix} \sin\delta\cos\phi \\ -\sin\delta\sin\phi \\ \cos\delta \end{bmatrix}$$
- **Ecuación de Cauchy:** $\mathbf{t} = \boldsymbol{\sigma}\cdot\hat{n}$
- Descomposición: $\mathbf{t}_n = (\mathbf{t}\cdot\hat{n})\hat{n}$ y $\mathbf{t}_t = \mathbf{t} - \mathbf{t}_n$

**Aplicación directa:** Mecánica de fallas, estabilidad de taludes.  
**Convención:** $x$ = Este, $y$ = Norte, $z$ = Arriba.

### 📓 [Ángulo de Buzamiento de Falla como Función de la Fricción](Ángulo%20de%20buzamiento%20de%20falla%20como%20función%20de%20la%20fricción.ipynb)

**Qué aprenderás:** Predecir la orientación de fallas a partir del coeficiente de fricción.  
**Modelo:** Turcotte y Schubert (2002)
- Fallas inversas: $\beta = \frac{1}{2}\arctan(1/f_s)$
- Fallas normales: $\beta = 90° + \frac{1}{2}\arctan(-1/f_s)$

**Área:** Geomecánica y tectónica.

## Referencia Rápida de Ecuaciones

### Eigenvalores y eigenvectores
$$\boldsymbol{A}\mathbf{v} = \lambda\mathbf{v} \quad \Leftrightarrow \quad \det(\boldsymbol{A} - \lambda\boldsymbol{I}) = 0$$

### Esfuerzo cortante máximo
$$\tau_{max} = \frac{\sigma_1 - \sigma_3}{2}$$

### Transformación tensorial bajo rotación
$$\boldsymbol{\sigma}' = \mathbf{R}\,\boldsymbol{\sigma}\,\mathbf{R}^T$$

### Ley de Hooke generalizada (notación indicial)
$$\sigma_{ij} = \lambda\,\delta_{ij}\,\varepsilon_{kk} + 2\mu\,\varepsilon_{ij}$$

### Ecuación de Cauchy (tracción en un plano)
$$\mathbf{t} = \boldsymbol{\sigma}\cdot\hat{n}$$

### Deformación volumétrica
$$\varepsilon_v = \text{tr}(\boldsymbol{\varepsilon}) = \varepsilon_{xx} + \varepsilon_{yy} + \varepsilon_{zz}$$

### Relaciones entre módulos elásticos
$$E = 2G(1+\nu) = 3K(1-2\nu) \qquad \lambda = \frac{E\nu}{(1+\nu)(1-2\nu)} \qquad \mu = \frac{E}{2(1+\nu)}$$
