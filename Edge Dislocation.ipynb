### Plotting cartesian stress components near an edge dislocation
fig_08_11 Pollard and Fletcher Fundamentals of Structural Geology equations (8.41) - (8.43)

after Weertman & Weertman (1964)


```julia
b = 1e-6  # Burgers vector
mu = 30000
pr = 0.25
lm = (2 * mu * pr) / (1 - 2 * pr)
c1 = b / (2 * π)
c2 = 2 * mu * (mu + lm) / (2 * mu + lm)
```

```julia
X = range(-0.0001, stop=0.0001, length=100) .+ eps()
Y = range(-0.0001, stop=0.0001, length=100)
```

```julia
DEN = @. (X.^2 + Y.^2).^2 # you can pass 1d arrays for the axes and it knows how to broadcast them, 
                          # and you write 2d functions of the axes variables by broadcast operations
                          # so you don't need something like meshgrid in MATLAB
```

```julia
SXX = c1 * c2 .* ((Y .* (3 * X.^2 + Y.^2)) ./ DEN)
SYY = -c1 * c2 .* ((Y .* (X.^2 - Y.^2)) ./ DEN)
SXY = -c1 * c2 .* ((X .* (X.^2 - Y.^2)) ./ DEN)
```

```julia
T, R = atan.(Y, X), sqrt.(X.^2 .+ Y.^2)
SXX[R .< (5 * b)] .= NaN
SYY[R .< (5 * b)] .= NaN
SXY[R .< (5 * b)] .= NaN
```

```julia
plot(X,Y,SXX,xlabel="x/b", ylabel="y/b", zlabel="stress sxx")
```

```julia
plot(X,Y,SYY,xlabel="x/b", ylabel="y/b", zlabel="stress syy")
```

```julia
plot(X,Y,SXY,xlabel="x/b", ylabel="y/b", zlabel="stress sxy")
```
