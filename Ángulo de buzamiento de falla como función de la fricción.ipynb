# Cálculo del Ángulo de Echado como Función de la Fricción

El ángulo de echado (buzamiento) de fallas inversas y normales puede calcularse en función del coeficiente de fricción \$ f_s $. Según **Turcotte y Schubert (p. 637)**, la relación se expresa como:

### 🔹 Para fallas inversas:
\begin{equation}
\beta_1 = \frac{\tan^{-1}(1/f_s)}{2} \cdot \frac{180}{\pi}
\end{equation}

### 🔹 Para fallas normales:
\begin{equation}
\beta_2 = \frac{\tan^{-1}(-1/f_s)}{2} \cdot \frac{180}{\pi}
\end{equation}

Dado que  $ \beta $ se mide contra el esfuerzo máximo $ \sigma_1 $, la corrección para fallas normales es:

\begin{equation}
\beta_2 = 90 + \beta_2
\end{equation}


```julia
# Calcular el ángulo de buzamiento de falla como función de la fricción 
fs = 0:0.05:1.0 ; # Coeficiente de fricción
pi2 = 180 / π  # Conversión de radianes a grados

# Ecuaciones de Turcotte y Schubert, pág. 637
beta1 = (atan.(1 ./ fs) ./ 2) .* pi2  # Fallas inversas
beta2 = (atan.(-1 ./ fs) ./ 2) .* pi2  # Fallas normales

# Corrección para fallas normales (medida contra sigma1)
beta2 .+= 90  
```

```julia
#  Graficar los resultados
plot(fs, beta1, label="Fallas inversas", color=:blue, linewidth=2)
plot!(fs, beta2, label="Fallas normales", color=:red, linewidth=2)
xlabel!("Coeficiente de Fricción")
ylabel!("Ángulo de Echado o Buzamiento")
plot!(grid=true)
```
