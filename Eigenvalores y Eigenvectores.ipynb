```python
import numpy as np

sigA=np.array ( [ [  4,  -2 ],
                  [  1,  1 ],
                   ])

print(sigA)
```

```python
L,v = np.linalg.eig(sigA)

print('Eigenvalues:')
print(L)
print()
print('Eigenvectors:')
print(v)
```

```python
#v = v.T  # Transpose the vectors if needed
#for vi in v:
  #  print(np.round(vi,3))
```

```python
vx = v[:, 0].reshape(2, 1) # Extract the first eigenvector and ensure it's a column vector
print(vx)
```

```python
# Compute sigA * vx
result = np.dot(sigA, vx)
print(result)
```

```python
2.68328157/3  # Confirms that the result is just the values of the eigenvector divided by the first eigenvalue
```
