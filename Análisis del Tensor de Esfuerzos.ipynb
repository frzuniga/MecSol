## **Análisis del Tensor de Esfuerzos**
En esta sesión, analizamos un **tensor de esfuerzos 3D**. Calcularemos los **esfuerzos principales**, sus **direcciones principales** y los visualizaremos en 3D.


```julia
# Cargar librerías necesarias
using LinearAlgebra
using Plots
```

## Definir el Tensor de Esfuerzos (Matriz Simétrica 3×3)

El **tensor de esfuerzos** en un sistema tridimensional se representa mediante una **matriz simétrica**:

\begin{equation}
\sigma =
\begin{bmatrix}
\sigma_{xx} & \tau_{xy} & \tau_{xz} \\
\tau_{xy} & \sigma_{yy} & \tau_{yz} \\
\tau_{xz} & \tau_{yz} & \sigma_{zz}
\end{bmatrix}
\end{equation}

Para este ejemplo, usamos el siguiente tensor de esfuerzos:

\begin{equation}
\sigma =
\begin{bmatrix}
10 & 2 & 3 \\
2 & 8 & 4 \\
3 & 4 & 6
\end{bmatrix}
\end{equation}

donde:

$( \sigma_{xx}, \sigma_{yy}, \sigma_{zz}) $ son los **esfuerzos normales** en las direcciones $( x, y, z )$.
$( \tau_{xy}, \tau_{xz}, \tau_{yz} )$ son los **esfuerzos cortantes** en los planos correspondientes.


```julia
# Definir un tensor de esfuerzos de ejemplo
σ = [10 2 3; 
     2 8 4; 
     3 4 6]
```

## **Tensor de Esfuerzos**
El tensor de esfuerzos en este ejemplo es:

```julia
# Calcular valores y vectores propios (esfuerzos principales y direcciones)
valores_propios, vectores_propios = eigen(σ)
```

##  **Esfuerzos Principales**
Los esfuerzos principales (valores propios del tensor de esfuerzos) son:

```julia
display(valores_propios)
```

## **Direcciones Principales**
Los vectores propios corresponden a las **direcciones principales de esfuerzo**:


```julia
display(vectores_propios)
```

```julia
# Función para graficar el tensor de esfuerzos en 3D
function graficar_tensor_esfuerzos(σ)
    
    ## 📌 **Visualización del Tensor de Esfuerzos**
    ## En la siguiente gráfica, se muestran los vectores de esfuerzo en 3D.
  
    x = [1, 0, 0]
    y = [0, 1, 0]
    z = [0, 0, 1]

    # Aplicar el tensor de esfuerzos a los ejes base
    fx = σ * x
    fy = σ * y
    fz = σ * z

    # Graficar los vectores de esfuerzo
    plot([0, fx[1]], [0, fx[2]], [0, fx[3]], label="σ_x", linewidth=3, arrow=:true, grid=true)
    plot!([0, fy[1]], [0, fy[2]], [0, fy[3]], label="σ_y", linewidth=3, arrow=:true, grid=true)
    plot!([0, fz[1]], [0, fz[2]], [0, fz[3]], label="σ_z", linewidth=3, arrow=:true, grid=true)

    title!("Visualización del Tensor de Esfuerzos")
    xlabel!("X")
    ylabel!("Y")
    zlabel!("Z")
    
end

# Llamar a la función de graficar
graficar_tensor_esfuerzos(σ)
```

```julia

function graficar_esfuerzos_principales(σ)
    """
    Grafica los esfuerzos principales en 3D.
    """

    # Calcular valores y vectores propios (esfuerzos principales y direcciones principales)
    valores_propios, vectores_propios = eigen(σ)

    println("\nEsfuerzos Principales: ", valores_propios)
    println("\nDirecciones Principales (Vectores Propios): ")
    display(vectores_propios)

    # Crear el gráfico 3D con los vectores propios
    plot()
    for i in 1:3
        plot!([0, vectores_propios[1, i]], 
              [0, vectores_propios[2, i]], 
              [0, vectores_propios[3, i]], 
              label="Eje Principal $i", linewidth=3, arrow=true, grid=true)
    end

    title!("Ejes Principales del Tensor de Esfuerzos")
    xlabel!("X")
    ylabel!("Y")
    zlabel!("Z")
end


# Llamar a la función corregida para graficar los esfuerzos principales
graficar_esfuerzos_principales(σ)
```
